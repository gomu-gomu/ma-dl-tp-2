## tape 1 : Introduction aux couches dun rseau

### Question conceptuelle

> Quest-ce quune couche dans un rseau de neurones artificiels ?

Une couche dans un rseau de neurones artificiels est une collection de neurones qui appliquent des transformations sur les donnes d'entre, en utilisant des poids, des biais, et une fonction d'activation. Les couches sont des lments fondamentaux qui structurent le rseau et permettent d'apprendre des relations entre les donnes.

> Dcrivez brivement le rle dune couche cache et dune couche de sortie.

Une couche cache se trouve entre la couche d'entre et la couche de sortie. Elle a pour rle principal de transformer les donnes d'entre en apprenant des caractristiques abstraites et des relations complexes.
La couche de sortie est la dernire couche du rseau et a pour rle de produire le rsultat final, qui correspond  la prdiction ou  la dcision du modle.

## tape 2 : Implmentation de la couche 1

### Implmentez la propagation avant pour la couche 1

> Crez une fonction Python qui calcule `Z(1)` et `A(1)`:

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def forward_couche_1(X, W1, b1):
    Z1 = np.dot(X, W1.T) + b1
    A1 = sigmoid(Z1)
    
    return Z1, A1
```

> Testez votre implmentation

```python
X = np.array([[0.5, 1.0]])
W1 = np.array([[0.2, 0.4], [0.3, 0.1]])
b1 = np.array([[0.1, -0.2]])

Z1, A1 = forward_couche_1(X, W1, b1)
Z1, A1
```

### Questions

> Pourquoi applique-t-on une fonction dactivation comme la sigmode aprs `Z(1)`?

On applique la fonction d'activation sigmode aprs `Z(1)` pour introduire de la non-linarit, ce qui permet au rseau de modliser des relations complexes entre les donnes. Elle normalise galement les sorties dans l'intervalle `[0, 1]`, ce qui facilite la stabilit des calculs et la convergence lors de l'entranement. De plus, la sigmode offre une interprtation probabiliste des rsultats, particulirement utile pour les tches de classification. Enfin, sa diffrentiabilit permet de calculer les gradients ncessaires  l'algorithme de rtropropagation, essentiel pour ajuster les paramtres du rseau.

> Que se passe-t-il si `W(1)` ou `b(1)` sont trs grands ?

Si `W(1)` ou `b(1)` sont trs grands, cela peut provoquer la saturation de la fonction sigmode, o les drives deviennent proches de zro, rendant la rtropropagation inefficace (problme de gradients nuls). De plus, cela peut entraner des sorties extrmes, une instabilit numrique et ralentir l'apprentissage. Pour viter cela, on initialise gnralement les poids avec de petites valeurs et les biais proches de zro, tout en normalisant les donnes pour stabiliser l'entranement.

## tape 4 : Vectorisation dun rseau de neurones

> Implmentez une fonction Python qui effectue les calculs vectoriss pour les deux 
couches :

```python
def forward_propagation_vectorized(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)
    
    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2)
    
    return Z1, A1, Z2, A2
```

> Testez cette fonction avec un lot de donnes `X` (par exemple, une matrice de dimension 
23, o chaque colonne reprsente un exemple) :

```python
X_v = np.array([[0.5, 0.2, 0.8], [1.0, 0.8, 0.4]])

Z1_v, A1_v, Z2_v, A2_v = forward_propagation_vectorized(X_v.T, W1, b1, W2, b2)
Z1_v, A1_v, Z2_v, A2_v
```
